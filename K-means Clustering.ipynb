```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
```

```python
from sklearn.cluster import KMeans
```

```python
#generation de donn√©es 
X, y=make_blobs(n_samples=100, centers=3,cluster_std=0.4, random_state=0)
plt.scatter(X[:,0],X[:,1],c='blue', marker='o', edgecolors='black')
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(X)
plt.scatter(X[:,0],X[:,1], c=model.predict(X))
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='r')
```

```python
model.cluster_centers_
```

```python
Y_hat=model.fit(X).labels_
Y_hat
```

```python
model.inertia_
```

```python
model.score(X,y)
```

```python
#elbow method
inertia=[]
K_range=range(1,20)
for k in K_range:
    model=KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)
plt.plot(K_range,inertia)
plt.xlabel('nombre de clusters')
plt.ylabel('cout du modele(inertia)')
#zone de crud
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X_iris = iris.data
kmeans_iris = KMeans(n_clusters=3)
kmeans_iris.fit(X_iris)
```

```python
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=kmeans_iris.labels_)
plt.scatter(kmeans_iris.cluster_centers_[:, 0], kmeans_iris.cluster_centers_[:, 1], c='r')
plt.xlabel('longeur')
plt.ylabel('Largeur ')

plt.show()
```

```python
kmeans_iris.cluster_centers_
```

```python
Y_hat=kmeans_iris.fit(X).labels_
Y_hat
```

```python
inertia=[]
K_range = range(1, 20)
for k in K_range:
    model = KMeans(n_clusters=k).fit(X_iris)
    inertia.append(model.inertia_)
plt.plot(K_range, inertia)
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie (cout du modele)')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
X, y=make_blobs(n_samples=50, centers=1, cluster_std=0.1, random_state=0)
X[-1,:]=np.array([2.25,5])
plt.scatter(X[:,0],X[:,1])
```

```python
model=IsolationForest(contamination=0.01)
model.fit(X)
plt.scatter(X[:,0],X[:,1],c=model.predict(X))
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
images=digits.images
X=digits.data
y=digits.target
plt.imshow(images[0])
```

```python
model=IsolationForest(random_state=0 ,contamination=0.02)
model.fit(X)
model.predict(X)
```

```python
model=IsolationForest(random_state=0, contamination=0.02)
model.fit(X)
outliers=model.predict(X)==-1
plt.figure(figsize=(12,3))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(images[outliers][i])
    plt.title(y[outliers][i])
```

```python
from sklearn.decomposition import PCA
model=PCA(n_components=2)
model.fit(X)
```

```python
x_pca=model.transform(X)
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
```

```python
x_pca.shape
```

```python
plt.figure()
plt.xlim(-30,30)
plt.ylim(-30,30)
for i in range(100):
    plt.text(x_pca[i,0],x_pca[i,1],str(y[i]))
```
